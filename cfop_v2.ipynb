```python
from selenium import webdriver # Automatiza a a web(Roda em segudo plano)
from selenium.webdriver.common.keys import Keys # Permite escrever textos no navegador
import pandas as pd
import numpy as np
from IPython.display import display_html

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Abre o navegador
driver = webdriver.Chrome('chromedriver',options=options)
driver.get("https://www.sefaz.pe.gov.br/Legislacao/Tributaria/Documents/Legislacao/Tabelas/CFOP.htm")
```

```python
dados = driver.find_elements_by_css_selector("body > div.WordSection1 > div:nth-child(20) > table")
#for row in dados:
#    print(row.text)
df= pd.read_html(driver.page_source, header=0)
```

```python
#Gerando DF para cada CFOP
df1 = df[1].head()
df2 = df[2].head()
df3 = df[3].head()
df4 = df[4].head()
df5 = df[5].head()
df6 = df[6].head()
```

```python
cfop = pd.concat([df1, df2,df3,df4,df5,df6])
```

```python
cfop.loc[:,'CFOP']= cfop['CFOP'].apply(lambda x: str(x).replace(',', '.'))
```
